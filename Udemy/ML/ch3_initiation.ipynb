12 Initiation à Matplotlib

```python
import matplotlib.pyplot as plt

#Allow to print out plots in jupyter (it can then be converted in HTML)
%matplotlib inline
```

```python
absc = [1,2,3,4]
ord =  [1,2,4,4]
plt.plot(absc,ord, 'o')
plt.title("Prix de stock")
plt.xlabel("Temps (s)")
plt.ylabel("Prix ($)")
plt.show()
```

```python
prix_microsoft = [1,3,2,4]
prix_apple = [2,4,4,3]

plt.plot(absc, prix_microsoft, label = "Microsoft")
plt.plot(absc, prix_apple, label = "Apple")

plt.title("Prix de stock")
plt.xlabel("Temps (s)")
plt.ylabel("Prix ($)")

plt.legend() #prints curbs labels
plt.show()
```
